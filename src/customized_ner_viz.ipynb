```python
import numpy as np
import pandas as pd
import os
import re
import glob
import spacy
from spacy.util import minibatch, compounding
from tqdm import tqdm, tqdm_notebook
from spacy import displacy
```

```python
IT_ner_ml = spacy.load('model/' + "IT_ner_model")
IT_job_2019 = pd.read_csv('../result/'  + 'IT_job_entity_2019.csv')
IT_job_2019.columns
```

```python
IT_entities = ['certification','diploma', 'experience_len', 'level','major', 'physical_req','remote','soft_skill', 'tech_skill', 'us_id_clearance']
rgb_colors = ['rgb(125, 231, 166)', 'rgb(125, 231, 218)', 'rgb(125, 138, 231)','rgb(166, 125, 231)','rgb(218, 125, 231)','rgb(231, 125, 138)','rgb(190, 231, 125)','rgb(231, 166, 125)','rgba(9, 111, 174, 0.6)','rgb(230, 226, 9)']
```

```python
sample_job_desc = IT_job_2019.loc[9,'job_description']
sen = IT_ner_ml(sample_job_desc)
IT_entities = [entity.upper() for entity in IT_entities]
colors = {IT_entities[i]: rgb_colors[i] for i in range(len(IT_entities))}
options = {"ents": IT_entities, "colors": colors}
displacy.render(sen, style='ent', jupyter=True, options=options)
```
